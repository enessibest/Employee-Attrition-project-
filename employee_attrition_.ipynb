```python
d4=d3.drop(['LastWorkingDate','Salary','Dateofjoining','Gender','City','MMM-YY','Education_Level'], axis = 1)
```

```python
d4.head()
```

```python
X = d4.drop(['Target'], axis=1)
Y = d4['Target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train,X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2)
```

```python
knn_mode = KNeighborsClassifier(n_neighbors=7)
knn_mode.fit (X_train, Y_train)
```

```python
pred= knn_mode.predict(X_test)
pred
```

```python
print('The accuracy of NB is {}%'.format(round(accuracy_score(pred,Y_test)*100,2)))
```
